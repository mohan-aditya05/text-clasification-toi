```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from requests import get
response=list()
for i in range(30):
    url = 'http://timesofindia.indiatimes.com/topic/hiv-aids/'+str(i)
    response.append(get(url))
    
```

```python
from bs4 import BeautifulSoup
html_soup=list()
for resp in response:
    html_soup.append(BeautifulSoup(resp.text, 'html.parser'))
```

```python
article_container = html_soup[0].find_all('li', class_ = 'article')
```

```python
first_article=article_container[0]
```

```python
links=list()
for article_container in html_soup:
    article_container=article_container.find_all('li',class_='article')
    for first_article in article_container:
        if(re.search("HIV",first_article.a.text)):
            links.append(first_article.a['href'])
    
```

```python
links
```

```python
response=list()
for link in links:
    url = 'http://timesofindia.indiatimes.com'+link
    response.append(get(url))
```

```python
html_soup=list()
for resp in response:
    if resp.status_code==200:
        html_soup.append(BeautifulSoup(resp.text, 'html.parser'))
```

```python
data=pd.DataFrame(columns=['Title','Time','Content'])
for article_container in html_soup:
    article=article_container.find_all('div',class_='main-content')
    if str(type(article[0].arttitle))=="<class 'NoneType'>":
        print('do nothing')
    else:
        content=article[0]
        title=content.arttitle.text
        time=content.time['datetime']
        main_content=content.find_all('div',class_='Normal')[0].text
        data=data.append({'Title':title,'Time':time,'Content':main_content},ignore_index=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.to_csv('data')
```
